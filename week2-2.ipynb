```python
%load_ext autoreload
%autoreload 2
import pandas as pd
%matplotlib inline
import edhec_risk_kit as erk
ind= erk.get_ind_returns()
er=erk.annualize_rets(ind["1996":"2000"],12)
cov=ind["1996":"2000"].cov()
```

```python
def portfolio_return(weights,returns):
    """
    Weights-> Returns
    """
    return weights.T @ returns #@symbolizes matrix multiplication. 

def portfolio_vol(weights,covmat):
    """
    Weights->volatility
    """
    return (weights.T @ covmat @ weights)**0.5   
```

## 2 Asset Frontier

```python
l=["Games","Fin"]
```

```python
n_points=20
weights= [np.array([w, 1-w]) for w in np.linspace(0,1,n_points)] #linspace gives returns equally spaced numbers between the assigned limits.
```

## Here we have 2 assets and we are generating various possible weights for the assets and then we will plot graph. Note: we use w, 1-w because sum of weights should be 1. 

```python
weights
```

```python
l
```

```python
rets=[erk.portfolio_return(w,er[l]) for w in weights]
vols=[erk.portfolio_vol(w,cov.loc[l,l])for w in weights]
import pandas as pd
ef= pd.DataFrame({"R":rets,"Vol":vols})
ef.plot.scatter(x="Vol",y="R")
```
